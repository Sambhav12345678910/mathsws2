```python
import math

x = []
y = []

i = -4*math.pi
while i <= 4*math.pi:
    x.append(i)
    y.append(math.sin(i))
    i += 0.1

plt.plot(x,y)
plt.title("y = sin(x)")
plt.grid()
plt.show()
```

```python
x = []
y = []

i = -3*math.pi
while i <= 3*math.pi:
    x.append(i)
    y.append(math.cos(i))
    i += 0.1

plt.plot(x,y)
plt.title("y = cos(x)")
plt.grid()
plt.show()
```

```python
import math
x = []
y = []
theta = 0
while theta <= 2*math.pi:
    x.append(4*math.cos(theta) + 2)
    y.append(4*math.sin(theta) + 2)
    theta += 0.1

plt.plot(x,y)
plt.gca().set_aspect('equal')
plt.grid()
plt.show()
```

```python
import math

xc, yc = [], []
theta = 0
while theta <= 2*math.pi:
    xc.append(4*math.cos(theta) + 2)
    yc.append(4*math.sin(theta) + 2)
    theta += 0.1

xe, ye = [], []
theta = 0
while theta <= 2*math.pi:
    xe.append(3*math.cos(theta) + 2)
    ye.append(5*math.sin(theta) + 2)
    theta += 0.1

plt.plot(xc, yc)
plt.plot(xe, ye)
plt.gca().set_aspect('equal')
plt.grid()
plt.show()
```
