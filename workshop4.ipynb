```python
import matplotlib.pyplot as plt

labels = ['Bears', 'Dolphins', 'Whales']
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

axes[0, 0].pie([8, 150, 80], labels=labels)
axes[0, 0].set_title("2017")

axes[0, 1].pie([54, 77, 54], labels=labels)
axes[0, 1].set_title("2018")

axes[0, 2].pie([93, 32, 100], labels=labels)
axes[0, 2].set_title("2019")

axes[1, 0].pie([116, 11, 76], labels=labels)
axes[1, 0].set_title("2020")

axes[1, 1].pie([137, 6, 93], labels=labels)
axes[1, 1].set_title("2021")

axes[1, 2].pie([184, 1, 72], labels=labels)
axes[1, 2].set_title("2022")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt


cars = ['Red', 'Silver', 'Blue', 'White', 'Green']
numbers = [123, 230, 78, 193, 12]

plt.figure()
plt.bar(cars, numbers)
plt.title("Car Colors Count")
plt.xlabel("Car Color")
plt.ylabel("Number Seen")
plt.show()
```

```python
plt.figure()
plt.bar(cars, numbers, width=0.3)
plt.title("Bar Chart with width = 0.3")
plt.show()

plt.figure()
plt.barh(cars, numbers)
plt.title("Horizontal Bar Chart")
plt.show()

colors = ['red', 'silver', 'blue', 'white', 'green']
plt.figure()
plt.bar(cars, numbers, color=colors)
plt.title("Bar Chart with Matching Colors")
plt.show()
```

```python
plt.figure()
plt.pie(numbers, labels=cars)
plt.title("Car Colors Distribution")
plt.show()
```

```python
plt.figure()
myexplode = [0.3, 0, 0, 0, 0]
plt.pie(numbers, labels=cars, explode=myexplode)
plt.title("Pie Chart with Explode")
plt.show()

plt.figure()
plt.pie(numbers, labels=cars, explode=myexplode, shadow=True)
plt.title("Pie Chart with Explode + Shadow")
plt.show()

plt.figure()
plt.pie(numbers, labels=cars)
plt.legend(title="Car Colours:")
plt.title("Pie Chart with Legend")
plt.show()

plt.figure()
mycolours = ["magenta", "green", "cyan", "red", "yellow"]
plt.pie(numbers, labels=cars, colors=mycolours)
plt.title("Pie Chart with Custom Colors")
plt.show()
```
