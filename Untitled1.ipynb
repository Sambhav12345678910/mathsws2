```python
def sumOfDivisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total

num = 7
while True:
    if sumOfDivisors(num) == num:
        print("Next perfect number:", num)
        break
    num += 1
```

```python
marks = [20, 30, 35, 55, 70, 80, 90]

mean = sum(marks) / len(marks)
n = len(marks)
marks.sort()
if n % 2 == 1:
    median = marks[n//2]
else:
    median = (marks[n//2 - 1] + marks[n//2]) / 2

print("Mean:", mean)
print("Median:", median)
```

```python
def decimal_to_binary(n):
    binary = []
    rem = 0
    while(n!=1):
      n = n // 2
      rem = rem % 2
      binary.append(rem)
      rem = n
    binary.append(1)
    binary.reverse()

    return binary

print(decimal_to_binary(123))
```

```python
def prodDigits(n):
    product = 1
    for digit in str(n):
        product *= int(digit)
    return product

print(prodDigits(123))  # 6
```

```python
def multiply(*args):
    result = 1
    for i in args:
        result *= i
    return result

# a
print(multiply(1,2,3,4,5))

# b
print(multiply(12,13,14))
```

```python
def capitals(**kwargs):
    print(kwargs)
    print()
    for (key, value) in kwargs.items():
        print(key, "->", value)

capitals(Nepal="Ktm", India="Delhi", China="Beijing", France="Paris")
```

```python
set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}
set3 = {10, 20, 30}

# a. Length of set1
print("Length of set1:", len(set1))

# b. Sum of set2
print("Sum of set2:", sum(set2))

# c. Union
print("Union:", set1 | set2)

# d. Intersection
print("Intersection:", set1 & set2)

# e. set1 - set2
print("set1 - set2:", set1 - set2)

# f. set2 - set1
print("set2 - set1:", set2 - set1)

# g. Subset check
print("Is set3 subset of set1:", set3.issubset(set1))

# h. Complement (U = 1 to 50)
U = set(range(1, 51))
print("Complement of set1:", U - set1)

# i. Check if 15 is present
print("Is 15 in set1:", 15 in set1)

# j. Add 60 to set1
set1.add(60)
print("After adding 60 to set1:", set1)

# k. Remove 60 from set2
set2.discard(60)
print("After removing 60 from set2:", set2)

# l. Clear set3
set3.clear()
print("After clearing set3:", set3)

# m. Max and Min in set2
print("Max in set2:", max(set2))
print("Min in set2:", min(set2))
```
